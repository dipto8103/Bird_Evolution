```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Mahika/Desktop/NMIMS/Sem 7/RL/project/ELEData/ELEData/TraitData/bl_Specimens_per_family.csv")
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.columns
```

```python
df.familyIDS.value_counts()
```

```python
metrics = [
    'Number_of_species_per_family',
    'Average_number_of_specimens_per_family',
    'Average_number_of_complete_specimens_per_family',
    'Portion_of_family_with_at_least_one_full_specimen',
    'Portion_of_complete_specimens_per_family',
    'Portion_of_family_with_at_least_four_specimens',
    'Portion_of_family_with_at_least_four_full_traits_measured_per_species',
    'Portion_of_family_with_at_least_four_main_traits_measured_per_species'
]

colors = ["#1a7cc3", '#ff7f0e', '#2ca02c', '#d62728',
          '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']

for index, row in df.iterrows():
    family_name = row['familyIDS']
    values = [row[col] for col in metrics]

    plt.figure(figsize=(12, 6))
    bars = plt.bar(metrics, values, color=colors)

    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            height,
            f'{height:.2f}',
            ha='center',
            va='bottom',
            fontsize=10,
            fontweight='bold'
        )

    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Value')
    plt.title(f'Metrics for Family: {family_name}')
    plt.tight_layout()
    plt.show()
```
